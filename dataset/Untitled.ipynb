```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

%pylab inline
```

```python
plt.figure(figsize=(12, 12))

n_samples = 300
random_state = 170
n_noises = 1000
x, y = make_blobs(n_samples=n_samples, centers=2, random_state=random_state)
np.random.uniform()

uni_x = np.reshape(np.random.uniform(-20, 20, n_noises), newshape=(n_noises, 1))
uni_y = np.reshape(np.random.uniform(-20, 20, n_noises), newshape=(n_noises, 1))
uni_data = np.append(uni_x, uni_y, axis=1)

x.shape
uni_data.shape
x = np.append(x, uni_data, axis=0)

#plt.subplot(221)
plt.scatter(x[:, 0], x[:, 1])
plt.title("Incorrect Number of Blobs")

plt.show()
```
